$v_{0} = 55$km/h = 15.27m/s

$a = -3$ m/s

distance from intersection = $x_{0}$

reaction time $\delta = 0.8$s

yellow interval $\tau = 3$s

Width of intersection $W = 45$m

Find out safe distance to stop from, and safe distance to continue from.

$\begin{align*}
& a(t) = \frac{dv}{dt} \\
& \int_{0}^{t}adt = \int_{v_{0}}^{v(t)}dv \\
& at = v(t) - v_{0} \\
& v(t) = v_{0} + at \\
& v(t) = \frac{dx}{dt} \\
& \int_{x_{0}}^{x(t)}dx = \int_{0}^{t}(v_{0} + at)dt \\
& x(t) - x_{0} = v_{0}t + a\frac{t^2}{2} \\
& x(t) = x_{0} + v_{0}t + a\frac{t^2}{2}
\end{align*}$


Now we use parameters. What is driver's position after reaction time $t = .8s$, here there is no acceleration yet, just constant velocity for .8s. And we know that $x_{0}$ is negative .

$x(.8) = - x_{0} + 15.27(.8) = -x_{0} + 12.216$

This is the position from which she must continue, or break.
Yellow time is 3s. So if she decides to go, we find the final position after 3s with no breaking, using the new position as our starting point. 

$x(3) = (-x{0} + 12.216) + 15.27(3) = -x_{0} + 58.025$

So if she continues, using the width of the intersection, we find the necessary value of $x_{0}$ 

$x_{0} = 58.026m - 45m = 13.026m$ 

So $x_{0}$ must be greater than 13.026m.

If she decides to break, we can look at two things: How long it takes to fully stop, and then where she would end up after stopping.

Since we have $at = v(t) - v_{0}$ We set v(t) = 0, since this is actually our final velocity, which we could have denoted during the integration.
$-3t = -15.27$ So $t = 5.09$

t here is larger than the time interval that the light is yellow for, so we can look now at where the driver ends up after the breaking:

$x(5.09) = (x_{0} + 12.216) + 15.27(5.09) - 3(\frac{5.09^2}{2})$

$x = (x_{0} + 51.078)$

This means $x_{0}$ must be at least 51.078m to fully stop before the intersection. 

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def x_break(x_0, v_0, a, t):
    return (x_0 + 12.216) + v_0*t + (a/2)*t**2
```

```python
def v_break(v_0, a, t):
    return v_0 + a*t
```

```python
def x_cont(x_0, v_0, t):
    return (x_0 + 12.216) + v_0*t
```

```python
def v_cont(v_0):
    return v_0
```

```python
t = np.linspace(0,7,70)

plt.axis([0,7,-40,80])

plt.plot(t, x_break(-30,15.27,-3,t), 'r', label='Break')
plt.plot(t, x_cont(-30, 15.27,t))

plt.axvline(x=3, color='r', linestyle='--')
plt.fill_between([0,7], [45,45], color='black', alpha=0.3)


plt.xlabel('time (s)')
plt.ylabel('position (m/s)')
plt.title('Trajectory for $x_{0} = -30$')
```

```python
t = np.linspace(0,7,70)

plt.axis([0,7,-80,80])

plt.plot(t, x_break(-70,15.27,-3,t), 'r', label='Break')
plt.plot(t, x_cont(-70, 15.27,t))

plt.axvline(x=3, color='r', linestyle='--')
plt.fill_between([0,7], [45,45], color='black', alpha=0.3)


plt.xlabel('time (s)')
plt.ylabel('position (m/s)')
plt.title('Trajectory for $x_{0} = -70$')
```

```python
def safe_to_drive(x,t,tau,W):
    if x > W and t > tau:
        return 'safe'
    else: 
        return 'unsafe'
def safe_to_break(x,t):
    if x < 0:
        return 'safe'
    else:
        return 'unsafe'
```
